# Histogram examples in Matplotlib

See also the docs at http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.hist

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# random samples from a normal distribution
x = np.random.randn(100)
```

```python
# default histogram
plt.hist(x)
```

```python
# change the bin number, thought the default is usually okay
plt.hist(x, bins=12)
```

```python
# set the color
plt.hist(x, color='c')
```

```python
# set the histogram type
plt.hist(x, histtype='step')
```

```python
# change the linewidth, can also use shortcut lw=4
plt.hist(x, histtype='step', linewidth=4)
```

```python
# make multiple histograms on the same plot
large_x = np.random.randn(1000) + 2
plt.hist(large_x, histtype='step', linewidth=4)
plt.hist(x, histtype='step', linewidth=4)
```

```python
# default is y axis in counts, try normalizing so area = 1 (y axis is probability)
plt.hist(large_x, histtype='step', linewidth=4, normed=True)
plt.hist(x, histtype='step', linewidth=4, normed=True)
```

```python
# alternate style with transparency set to 0.5
plt.hist(large_x, alpha=0.5, linewidth=4, normed=True)
plt.hist(x, alpha=0.5, linewidth=4, normed=True)
```

## 2d histogram example

```python
# 2d histogram example
x, y = np.random.multivariate_normal(mean=[1, 2], cov=[[1, 0.5], [0.5, 1]], size=100000).T
counts, yedges, xedges, image = plt.hist2d(x, y, bins=50, normed=True)
plt.colorbar()
```

```python
# or plot contours
xx, yy = np.meshgrid(xedges[:-1], yedges[:-1])
plt.contour(xx, yy, counts)
plt.colorbar(image)
```
