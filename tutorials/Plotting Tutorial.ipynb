The first thing to do when plotting in python is to import the matplotlib library. This allows you to use all of the matplotlib commands. Also, when using ipython notebooks it is useful to inclue the line: %matplotlib inline so that you can see the plots easily as you code

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

Let's also go ahead and import numpy so that we can make simple arrays that we can plot

```python
import numpy as np
```

Now let's use numpy to make an array of values from negative pi to pi. This will be our x-array that we'll use for plotting. Let's also set up two y-arrays using simple numpy functions for sine and cosine. 

```python
X = np.linspace(-np.pi, np.pi, num=100)
C, S = np.cos(X), np.sin(X)
```

We now have arrays for x values from negative pi to pi (array X), y values that are the cosine of the x values (C array) and y values that are the sine of the x values (S array). Now that we have these arrays we can plot them using the matplotlib syntax below:

```python
plt.plot(X, C)
plt.plot(X, S)
```

Voila! You've created a simple plot in python! Using matplotlib there are many ways to change how your plot looks. If you have enough patience you can make nearly any plot that you can think of. Here are just a few examples of how to change a plot

We can change the limits of the plot through the command: plt.xlim, plt.ylim. Try changing the limits of the plot yourself!

```python
plt.plot(X, C)
plt.plot(X, S)
plt.xlim(-np.pi,np.pi)
plt.ylim(-1.2,1.2)
```

We can further change the color of the lines, the width of the lines, and the style of the lines. Try googling other matplotlib linestyles and changing this yourself!

```python
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-.")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-.")
```

You can also easily add a legend to your plot! Try changing the labels on the legend and the location of the legend (you might have to google to see what words you have to tell the plt.legend() command).

```python
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-", label="sine")

plt.legend(loc='upper left')
```

Now you might want to label the axes of your plot and give it a title. Try changing what these labels are!

```python
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-", label="sine")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Sine and Cosine')


plt.legend(loc='upper left')
```

You can also plot a scatter plot instead of a continuous line! Or plot both at once! Try playing with these settings now and maybe adding your own!

```python
X_scatter = np.linspace(-np.pi, np.pi, num=10)
C_scatter, S_scatter = np.cos(X_scatter), np.sin(X_scatter)

plt.plot(X, C, color="blue", linewidth=.5, linestyle="--", label="cosine")
plt.plot(X, S, color="red",  linewidth=.5, linestyle="--", label="sine")
plt.scatter(X_scatter, C_scatter, color="blue")
plt.scatter(X_scatter, S_scatter, color="red")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Sine and Cosine')


plt.legend(loc='upper left')
```
