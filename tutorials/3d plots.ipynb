```python
%matplotlib notebook
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
# mock some data
mean = np.array([3, 4, 5])
cov=np.array([[0.5, 0.6, 0.5],
              [0.3, 0.5, 0.5],
              [0.5, 0.5, 0.9]])
x, y, z = np.random.multivariate_normal(mean=mean, cov=cov, size=500).T
points = np.vstack([x, y, z]).T
dist = np.sqrt(np.sum((points - mean) **2, axis=1))

# make the 3d axes, works similar to pyplot,
# but there are some frustrating differences
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=dist)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
